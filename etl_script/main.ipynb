```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
dim_date = df[['Order Date']].drop_duplicates().copy()
dim_date['date_key'] = dim_date['Order Date'].astype('int64') // 10**9  
dim_date['full_date'] = dim_date['Order Date']
dim_date['day_of_week'] = dim_date['Order Date'].dt.day_name()
dim_date['month'] = dim_date['Order Date'].dt.month
dim_date['quarter'] = dim_date['Order Date'].dt.quarter
dim_date['year'] = dim_date['Order Date'].dt.year
dim_date = dim_date[['date_key', 'full_date', 'day_of_week', 'month', 'quarter', 'year']]
```

```python
df['date_key'] = df['Order Date'].astype('int64') // 10**9
```

```python
df.rename(columns={
    'Country': 'country',
    'City': 'city',
    'State': 'state',
    'Postal Code': 'postal_code',
    'Region': 'region',
    'Latitude': 'latitude',
    'Longitude': 'longitude',
    'location_key': 'location_key',
    'Customer ID' : 'customer_id',
    'Customer Name' :'customer_name',
    'Product ID': 'product_id',
    'Category' : 'category',
    'Sub-Category' : 'sub_category',
    'Product Name' : 'product_name',
    'Sales' : 'sales',
    'Quantity' : 'quantity',
    'Discount' : 'discount',
    'Profit' :'profit',
    'Stock' : 'stock',
    'Segment' : 'segment',
    'Ship Mode' : 'ship_mode',
    'Order ID' : 'order_id',




}, inplace=True)


```

```python
dim_location = df[['country', 'city', 'state', 'postal_code', 'region', 'latitude', 'longitude']].drop_duplicates().reset_index(drop=True)
dim_location['location_key'] = dim_location.index + 1


df = df.merge(dim_location, on=['country', 'city', 'state', 'postal_code', 'region', 'latitude', 'longitude'], how='left')
```

```python
dim_customer = df[['customer_id', 'customer_name', 'segment', 'location_key']].drop_duplicates().reset_index(drop=True)
df = df.merge(dim_customer, on='customer_id', how='left')
```

```python
dim_product = df[['product_id', 'product_name', 'category', 'sub_category']].drop_duplicates().reset_index(drop=True)
df = df.merge(dim_product, on='product_id', how='left')
```

```python
dim_ship_mode = df[['ship_mode']].drop_duplicates().reset_index(drop=True)
dim_ship_mode['ship_mode_key'] = dim_ship_mode.index + 1
df = df.merge(dim_ship_mode, on='ship_mode', how='left')
```

```python
fact_sales = df[['order_id', 'date_key', 'customer_id', 'product_id', 'ship_mode_key', 'sales', 'quantity', 'discount', 'profit']].drop_duplicates()
fact_sales.columns = ['order_id', 'order_date_key', 'customer_id', 'product_id', 'ship_mode_key', 'sales', 'quantity', 'discount', 'profit']
```

Load

```python
from sqlalchemy import create_engine, text
from sqlalchemy.exc import SQLAlchemyError
```

```python

engine = create_engine("postgresql+psycopg2://postgres:root@localhost:5432/carrefour")
try:
    with engine.connect() as conn:
        conn.execute(text("SELECT 1"))  
    print("Koneksi ke database berhasil!")
except SQLAlchemyError as e:
    print("Gagal koneksi ke database:")
    print(e)
```

```python

create_dim_date = """
CREATE TABLE IF NOT EXISTS dim_date (
    date_key INT PRIMARY KEY,
    full_date DATE NOT NULL,
    day_of_week VARCHAR(10),
    month INT,
    quarter INT,
    year INT
);
"""

create_dim_location = """
CREATE TABLE IF NOT EXISTS dim_location (
    location_key SERIAL PRIMARY KEY,
    country TEXT,
    city TEXT,
    state TEXT,
    postal_code INT,
    region TEXT,
    latitude DOUBLE PRECISION,
    longitude DOUBLE PRECISION
);
"""

create_dim_customer = """
CREATE TABLE IF NOT EXISTS dim_customer (
    customer_id TEXT PRIMARY KEY,
    customer_name TEXT,
    segment TEXT,
    location_key INT NOT NULL REFERENCES dim_location(location_key)
);
"""

create_dim_product = """
CREATE TABLE IF NOT EXISTS dim_product (
    product_id TEXT PRIMARY KEY,
    product_name TEXT,
    category TEXT,
    sub_category TEXT
);
"""

create_dim_ship_mode = """
CREATE TABLE IF NOT EXISTS dim_ship_mode (
    ship_mode_key SERIAL PRIMARY KEY,
    ship_mode TEXT UNIQUE NOT NULL
);
"""

create_fact_sales = """
CREATE TABLE IF NOT EXISTS fact_sales (
    order_id TEXT PRIMARY KEY,
    order_date_key INT NOT NULL REFERENCES dim_date(date_key),
    customer_id TEXT NOT NULL REFERENCES dim_customer(customer_id),
    product_id TEXT NOT NULL REFERENCES dim_product(product_id),
    ship_mode_key INT NOT NULL REFERENCES dim_ship_mode(ship_mode_key),
    sales NUMERIC,
    quantity INT,
    discount NUMERIC,
    profit NUMERIC
);
"""

with engine.begin() as conn:
    conn.execute(text(create_dim_date))
    conn.execute(text(create_dim_location))
    conn.execute(text(create_dim_customer))
    conn.execute(text(create_dim_product))
    conn.execute(text(create_dim_ship_mode))
    conn.execute(text(create_fact_sales))

print("Tables created successfully")
```

```python
try:
    with engine.begin() as conn:  
        print("Cek koneksi ke DB...")
        conn.execute(text("SELECT 1"))
        print("Koneksi berhasil!")

        print("Membuat tabel dim_date...")
        conn.execute(text(create_dim_date))
        print("Tabel dim_date berhasil dibuat atau sudah ada.")

        print("Cek tabel di DB:")
        result = conn.execute(text("""
            SELECT table_name FROM information_schema.tables 
            WHERE table_schema = 'public';
        """))
        tables = [row[0] for row in result]
        print("Tabel yang ada:", tables)

except SQLAlchemyError as e:
    print("Error saat koneksi atau pembuatan tabel:")
    print(e)
```

```python
from sqlalchemy import text

with engine.begin() as conn:
    conn.execute(text("TRUNCATE TABLE fact_sales, dim_customer, dim_product, dim_ship_mode, dim_location, dim_date RESTART IDENTITY CASCADE"))

    # Insert dim_date
    dim_date_records = dim_date.to_dict(orient='records')
    conn.execute(
        text("""
        INSERT INTO dim_date (date_key, full_date, day_of_week, month, quarter, year)
        VALUES (:date_key, :full_date, :day_of_week, :month, :quarter, :year)
        ON CONFLICT (date_key) DO NOTHING
        """),
        dim_date_records
    )

    dim_product_records = dim_product.to_dict(orient='records')
    conn.execute(
        text("""
        INSERT INTO dim_location (country, city, state, postal_code, region, latitude, longitude)
        VALUES (:country, :city, :state, :postal_code, :region, :latitude, :longitude)
        ON CONFLICT DO NOTHING
        """),
        dim_location_records
    )

    dim_customer_records = dim_customer.to_dict(orient='records')
    conn.execute(
        text("""
        INSERT INTO dim_customer (customer_id, customer_name, segment, location_key)
        VALUES (:customer_id, :customer_name, :segment, :location_key)
        ON CONFLICT (customer_id) DO NOTHING
        """),
        dim_customer_records
    )

    dim_product_records = dim_product.to_dict(orient='records')

    conn.execute(
        text("""
        INSERT INTO dim_product (product_id, product_name, category, sub_category)
        VALUES (:product_id, :product_name, :category, :sub_category)
        ON CONFLICT (product_id) DO NOTHING
        """),
        dim_product_records
    )

    dim_ship_mode_records = dim_ship_mode.drop(columns=['ship_mode_key']).to_dict(orient='records')
    conn.execute(
        text("""
        INSERT INTO dim_ship_mode (ship_mode)
        VALUES (:ship_mode)
        ON CONFLICT (ship_mode) DO NOTHING
        """),
        dim_ship_mode_records
    )

    fact_sales_records = fact_sales.to_dict(orient='records')
    conn.execute(
        text("""
        INSERT INTO fact_sales (order_id, order_date_key, customer_id, product_id, ship_mode_key, sales, quantity, discount, profit)
        VALUES (:order_id, :order_date_key, :customer_id, :product_id, :ship_mode_key, :sales, :quantity, :discount, :profit)
        ON CONFLICT (order_id) DO NOTHING
        """),
        fact_sales_records
    )

print("Data inserted successfully!")
```
